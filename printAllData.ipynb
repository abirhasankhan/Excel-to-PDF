```python
import pandas as pd
from fpdf import FPDF
from google.colab import files

# Function to upload the Excel file in Colab
def upload_excel():
    uploaded = files.upload()  # This will open a file upload dialog in Colab
    if not uploaded:
        print("No file uploaded. Exiting...")
        exit()
    excel_path = list(uploaded.keys())[0]  # Get the uploaded file path
    return excel_path

# Function to create a table-like PDF with column splitting for wide tables
def save_pdf(data, output_filename, sheet_name):
    pdf = FPDF(orientation='L', unit='mm', format='A4')  # Landscape orientation
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    # Set title
    pdf.set_font("Arial", style='B', size=14)
    pdf.cell(0, 10, f"Sheet: {sheet_name}", ln=True, align='C')
    pdf.ln(10)

    # Maximum number of columns per page
    max_columns_per_page = 8
    total_columns = len(data.columns)

    # Split columns into chunks
    column_chunks = [data.columns[i:i + max_columns_per_page] for i in range(0, total_columns, max_columns_per_page)]

    # Process each column chunk
    for chunk_index, col_chunk in enumerate(column_chunks):
        if chunk_index > 0:
            pdf.add_page()  # Add a new page for each chunk

        pdf.set_font("Arial", style='B', size=10)
        col_width = 277 / len(col_chunk)  # Distribute columns evenly across the page

        # Add headers
        for header in col_chunk:
            pdf.cell(col_width, 10, str(header), border=1, align='C')
        pdf.ln()

        # Add rows
        pdf.set_font("Arial", size=10)
        for row in data[col_chunk].itertuples(index=False):
            for value in row:
                pdf.cell(col_width, 8, str(value), border=1, align='C')
            pdf.ln()

    # Save the PDF
    pdf.output(output_filename)
    print(f"PDF saved as {output_filename}")

# Upload the Excel file
excel_path = upload_excel()
xls = pd.ExcelFile(excel_path)

# Display sheet names and tables, then ask user for input
print("Available sheets:", xls.sheet_names)
for sheet_name in xls.sheet_names:
    data = xls.parse(sheet_name)
    print(f"\nSheet Name: {sheet_name}")
    print(data.head())  # Display the first few rows of each sheet (you can adjust this as needed)

# Ask the user for which sheets to export
user_input = input("\nEnter sheet names separated by commas or type 'all' to export all sheets: ")

sheet_names = xls.sheet_names if user_input.lower() == 'all' else [name.strip() for name in user_input.split(",")]

# Process each sheet and save as PDF
for sheet_name in sheet_names:
    try:
        data = xls.parse(sheet_name)
        output_pdf = f"{excel_path.rsplit('.', 1)[0]}_{sheet_name}.pdf"
        save_pdf(data, output_pdf, sheet_name)

        # Allow the user to download the PDF after it's generated
        files.download(output_pdf)

    except Exception as e:
        print(f"Error processing sheet {sheet_name}: {e}")
```
